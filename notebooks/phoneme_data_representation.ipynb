```python
import numpy as np
import pandas as pd
```

```python
cons_pl_npy = np.load("./data/consonants_plumonic.npy", allow_pickle=True)
cons_npl_npy = np.load("./data/consonants_non_plumonic.npy", allow_pickle=True)
cons_coart_npy = np.load("./data/consonants_coarticulated.npy", allow_pickle=True)
vowels_npy = np.load("./data/vowels.npy", allow_pickle=True)


cons_pl_npy.shape, cons_npl_npy.shape, cons_coart_npy.shape, vowels_npy.shape
```

```python
cons_pl_tbl_df = pd.read_csv(
    "./data/consonants_plumonic.csv",
    dtype=str,
    index_col=[0],
    keep_default_na=False,
    na_values=["-1"],
)
cons_pl_tbl_df
```

```python
cons_npl_tbl_df = pd.read_csv(
    "./data/consonants_non_plumonic.csv",
    dtype=str,
    index_col=[0],
    keep_default_na=False,
    na_values=["-1"],
)
cons_npl_tbl_df
```

```python
cons_coart_tbl_df = pd.read_csv(
    "./data/consonants_coarticulated.csv",
    dtype=str,
    index_col=[0],
    keep_default_na=False,
    na_values=["-1"],
)
cons_coart_tbl_df
```

```python
vowels_tbl_df = pd.read_csv(
    "./data/vowels.csv",
    dtype=str,
    index_col=[0],
    keep_default_na=False,
    na_values=["-1"],
)
vowels_tbl_df
```

```python
def map_phoneme(diacritics, handled, diacritic=False):
    def mapper(p):
        symbol, suffix = (p[0:-1], p[-1]) if (isinstance(p, str) and len(p) > 1 and p[-1] in diacritics) else (p, None)
        if diacritic:
            return suffix
        else:
            return symbol

    return mapper


def map_variants(diacritics, handled, tbl):
    symbol_tbl = cons_pl_tbl_df.map(
        map_phoneme(diacritics, handled),
    )
    suffix_tbl = cons_pl_tbl_df.map(
        map_phoneme(diacritics, handled, diacritic=True),
    )
    return symbol_tbl, suffix_tbl
```

```python
ipa_diacritics_df = pd.read_csv(
    "./data/ipa_diacritics.csv",
    dtype=str,
)
ipa_diacritics_df["bin"] = ipa_diacritics_df.index.map(
    lambda i: format(i, "b").zfill(6)
)
ipa_diacritics_df.head(5)
```

```python
diacritics = set(ipa_diacritics_df.suffix)
handled = set(
    np.concatenate(
        [
            cons_pl_tbl_df.stack().replace("", np.nan).dropna().to_numpy(),
            cons_npl_tbl_df.stack().replace("", np.nan).dropna().to_numpy(),
            cons_coart_tbl_df.stack().replace("", np.nan).dropna().to_numpy(),
            vowels_tbl_df.stack().replace("", np.nan).dropna().to_numpy(),
        ]
    )
)

cons_pl_symbol_tbl_df, cons_pl_suffix_tbl_df = map_variants(diacritics, handled, cons_pl_tbl_df)
cons_npl_symbol_tbl_df, cons_npl_suffix_tbl_df = map_variants(diacritics, handled, cons_npl_tbl_df)
cons_coart_symbol_tbl_df, cons_coart_suffix_tbl_df = map_variants(diacritics, handled, cons_coart_tbl_df)
vowels_symbol_tbl_df, vowels_suffix_tbl_df = map_variants(diacritics, handled, vowels_tbl_df)

print(
    "cons_pl_symbol_tbl_df",
    cons_pl_symbol_tbl_df.stack().replace("", np.nan).dropna().unique(),
)
print(
    "cons_npl_symbol_tbl_df",
    cons_npl_symbol_tbl_df.stack().replace("", np.nan).dropna().unique(),
)
print(
    "cons_coart_symbol_tbl_df",
    cons_coart_symbol_tbl_df.stack().replace("", np.nan).dropna().unique(),
)
print(
    "vowels_symbol_tbl_df",
    vowels_symbol_tbl_df.stack().replace("", np.nan).dropna().unique(),
)
```
